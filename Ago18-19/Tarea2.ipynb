```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
```

### Los Datos Obtenidos

Son datos de los puntos de los distintos pokemones. 

```python
data = pd.read_csv('Pokemon.csv')
```

```python
plt.scatter(data['Attack'], data['Total'], color = 'black', s = 10)
plt.title('Attack vs Total Points of Pokemons')
plt.xlabel('Points of Attack')
plt.ylabel('Total Points');
```

```python
def min_cuadrados(x, y):
    N = len(x)
    m = (np.sum(x*y) - np.sum(x)*np.sum(y)/N)/(np.sum(x*x) - (np.sum(x)**2)/N)
    b = (np.sum(y) - m*np.sum(x))/N
    
    return (m, b)

x = data['Attack']
y = data['Total']

slope, b0 = min_cuadrados(x, y)

x1 = np.linspace(np.min(x), np.max(x), 250)
```

```python
plt.scatter(data['Attack'], data['Total'], color = 'black', s = 10, label = 'Data')
plt.plot(x1, slope*x1+ b0, color = 'red', ls = '--')
plt.title('Defense vs Attack of Pokemons')
plt.xlabel('Points of Defense')
plt.ylabel('Points of Attack');
```
